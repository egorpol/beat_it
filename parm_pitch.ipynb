```python
import librosa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Import external libraries
import crepe
import parselmouth

# Define pitch analysis functions with pre_window
def analyze_pitch_yin(y, sr, cue_time, pre_window=0.1, window_duration=0.75):
    """
    Analyze pitch using librosa.yin with a pre-window.
    """
    start_time = max(cue_time - pre_window, 0)
    end_time = cue_time + window_duration
    start_sample = int(start_time * sr)
    end_sample = int(end_time * sr)
    end_sample = min(end_sample, len(y))

    y_segment = y[start_sample:end_sample]

    if np.all(y_segment == 0):
        return np.nan

    try:
        f0 = librosa.yin(
            y_segment,
            fmin=librosa.note_to_hz('C2'),
            fmax=librosa.note_to_hz('C7'),
            sr=sr
        )
        f0 = f0[f0 > 0]
        if len(f0) == 0:
            return np.nan
        return np.mean(f0)
    except Exception as e:
        print(f"YIN analysis failed at {cue_time}s: {e}")
        return np.nan

def analyze_pitch_piptrack(y, sr, cue_time, pre_window=0.1, window_duration=0.75):
    """
    Analyze pitch using librosa.piptrack with a pre-window.
    """
    start_time = max(cue_time - pre_window, 0)
    end_time = cue_time + window_duration
    start_sample = int(start_time * sr)
    end_sample = int(end_time * sr)
    end_sample = min(end_sample, len(y))

    y_segment = y[start_sample:end_sample]

    if np.all(y_segment == 0):
        return np.nan

    try:
        S = librosa.stft(y_segment)
        magnitude, phase = librosa.magphase(S)
        pitches, magnitudes = librosa.piptrack(
            S=magnitude,
            sr=sr,
            fmin=librosa.note_to_hz('C2'),
            fmax=librosa.note_to_hz('C7')
        )
        pitch_values = []
        for i in range(pitches.shape[1]):
            index = magnitudes[:, i].argmax()
            pitch = pitches[index, i]
            if pitch > 0:
                pitch_values.append(pitch)
        if not pitch_values:
            return np.nan
        return np.mean(pitch_values)
    except Exception as e:
        print(f"PIPTRACK analysis failed at {cue_time}s: {e}")
        return np.nan

def analyze_pitch_crepe_func(y, sr, cue_time, pre_window=0.1, window_duration=0.75):
    """
    Analyze pitch using CREPE with a pre-window.
    """
    start_time = max(cue_time - pre_window, 0)
    end_time = cue_time + window_duration
    start_sample = int(start_time * sr)
    end_sample = int(end_time * sr)
    end_sample = min(end_sample, len(y))

    y_segment = y[start_sample:end_sample]

    if np.all(y_segment == 0):
        return np.nan

    try:
        # Resample to 16000 Hz as required by CREPE
        y_resampled = librosa.resample(y_segment, orig_sr=sr, target_sr=16000)
        time, frequency, confidence, activation = crepe.predict(
            y_resampled,
            16000,
            viterbi=True
        )
        frequency = frequency[confidence > 0.5]
        if len(frequency) == 0:
            return np.nan
        return np.mean(frequency)
    except Exception as e:
        print(f"CREPE analysis failed at {cue_time}s: {e}")
        return np.nan

def analyze_pitch_parselmouth_func(y, sr, cue_time, pre_window=0.1, window_duration=0.75):
    """
    Analyze pitch using Parselmouth (Praat) with a pre-window.
    """
    try:
        sound = parselmouth.Sound(y, sr)
        start_time = max(cue_time - pre_window, 0)
        end_time = cue_time + window_duration
        end_time = min(end_time, sound.duration)
        windowed_sound = sound.extract_part(
            from_time=start_time,
            to_time=end_time,
            preserve_times=True
        )
        pitch = windowed_sound.to_pitch(
            time_step=0.01,
            pitch_floor=librosa.note_to_hz('C2'),
            pitch_ceiling=librosa.note_to_hz('C7')
        )
        pitch_values = pitch.selected_array['frequency']
        pitch_values = pitch_values[pitch_values > 0]
        if len(pitch_values) == 0:
            return np.nan
        return np.mean(pitch_values)
    except Exception as e:
        print(f"Parselmouth analysis failed at {cue_time}s: {e}")
        return np.nan

# Helper function to convert note names to MIDI numbers
def note_to_midi(note):
    if pd.isna(note):
        return np.nan
    else:
        try:
            return librosa.note_to_midi(note)
        except:
            return np.nan



# Load audio
audio_path = 'parm2_mono.aif'  # Replace with your audio file path
y, sr = librosa.load(audio_path, sr=None)

# Define window durations
pre_window = 0.1      # seconds before the cue point
window_duration = 0.75 # seconds after the cue point

# Analyze pitch using YIN
df.loc[:, 'PITCH_HZ_YIN'] = df['TIME'].apply(
    lambda t: analyze_pitch_yin(y, sr, t, pre_window, window_duration)
)

# Analyze pitch using PIPTRACK
df.loc[:, 'PITCH_HZ_PIPTRACK'] = df['TIME'].apply(
    lambda t: analyze_pitch_piptrack(y, sr, t, pre_window, window_duration)
)

# Analyze pitch using CREPE
df.loc[:, 'PITCH_HZ_CREPE'] = df['TIME'].apply(
    lambda t: analyze_pitch_crepe_func(y, sr, t, pre_window, window_duration)
)

# Analyze pitch using Parselmouth (Praat)
df.loc[:, 'PITCH_HZ_PARS'] = df['TIME'].apply(
    lambda t: analyze_pitch_parselmouth_func(y, sr, t, pre_window, window_duration)
)

# Convert frequency to note names for each method
df.loc[:, 'PITCH_NOTE_YIN'] = df['PITCH_HZ_YIN'].apply(
    lambda f: librosa.hz_to_note(f) if not np.isnan(f) else np.nan
)
df.loc[:, 'PITCH_NOTE_PIPTRACK'] = df['PITCH_HZ_PIPTRACK'].apply(
    lambda f: librosa.hz_to_note(f) if not np.isnan(f) else np.nan
)
df.loc[:, 'PITCH_NOTE_CREPE'] = df['PITCH_HZ_CREPE'].apply(
    lambda f: librosa.hz_to_note(f) if not np.isnan(f) else np.nan
)
df.loc[:, 'PITCH_NOTE_PARS'] = df['PITCH_HZ_PARS'].apply(
    lambda f: librosa.hz_to_note(f) if not np.isnan(f) else np.nan
)

# Convert note names to MIDI numbers for each method
df.loc[:, 'MIDI_NOTE_YIN'] = df['PITCH_NOTE_YIN'].apply(note_to_midi)
df.loc[:, 'MIDI_NOTE_PIPTRACK'] = df['PITCH_NOTE_PIPTRACK'].apply(note_to_midi)
df.loc[:, 'MIDI_NOTE_CREPE'] = df['PITCH_NOTE_CREPE'].apply(note_to_midi)
df.loc[:, 'MIDI_NOTE_PARS'] = df['PITCH_NOTE_PARS'].apply(note_to_midi)

# Display the updated DataFrame
display(df)

# Plotting all pitch estimates
plt.figure(figsize=(12, 8))
plt.plot(df['TIME'], df['PITCH_HZ_YIN'], marker='o', linestyle='-', label='YIN Pitch')
plt.plot(df['TIME'], df['PITCH_HZ_PIPTRACK'], marker='x', linestyle='--', label='PIPTRACK Pitch')
plt.plot(df['TIME'], df['PITCH_HZ_CREPE'], marker='s', linestyle=':', label='CREPE Pitch')
plt.plot(df['TIME'], df['PITCH_HZ_PARS'], marker='d', linestyle='-.', label='Parselmouth Pitch')
plt.xlabel('Time (s)')
plt.ylabel('Average Pitch (Hz)')
plt.title('Pitch Analysis at Cue Points Using Multiple Methods')
plt.legend()
plt.grid(True)
plt.show()

# Piano Roll Plot
plt.figure(figsize=(12, 8))

# Plot for each method
plt.scatter(df['TIME'], df['MIDI_NOTE_YIN'], marker='o', label='YIN')
plt.scatter(df['TIME'], df['MIDI_NOTE_PIPTRACK'], marker='x', label='PIPTRACK')
plt.scatter(df['TIME'], df['MIDI_NOTE_CREPE'], marker='s', label='CREPE')
plt.scatter(df['TIME'], df['MIDI_NOTE_PARS'], marker='d', label='Parselmouth')

# Formatting the y-axis to show note names
min_midi = int(df[['MIDI_NOTE_YIN', 'MIDI_NOTE_PIPTRACK', 'MIDI_NOTE_CREPE', 'MIDI_NOTE_PARS']].min().min()) - 1
max_midi = int(df[['MIDI_NOTE_YIN', 'MIDI_NOTE_PIPTRACK', 'MIDI_NOTE_CREPE', 'MIDI_NOTE_PARS']].max().max()) + 1
yticks = np.arange(min_midi, max_midi + 1)
# yticks = np.arange(36, 85)  # MIDI notes from C2 to C6
ylabels = [librosa.midi_to_note(midi) for midi in yticks]
plt.yticks(yticks, ylabels)

plt.xlabel('Time (s)')
plt.ylabel('Note')
plt.title('Piano Roll of Estimated Notes at Cue Points')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Step 1: Load CSV from URL
url = 'https://raw.githubusercontent.com/egorpol/beat_it/refs/heads/main/csv/parm_synth_pitch_manual.csv'
manual_df = pd.read_csv(url)

# # Step 2: Filter the DataFrame and create a copy to avoid SettingWithCopyWarning
# df = df4[df4['LABEL'] == 's'].copy()

manual_df = manual_df.rename(columns={'LABEL': 'LABEL_MANUAL'})

# Display the first few rows
display(manual_df)
```

```python
# Merge the manually tagged DataFrame with the automatically analyzed DataFrame
merged_df = pd.merge(df, manual_df, on='TIME', how='inner')

# Display the merged DataFrame
display(merged_df)
```

```python
import librosa
import numpy as np

# Define helper function to convert note names to MIDI numbers
def note_to_midi(note):
    if pd.isna(note):
        return np.nan
    else:
        try:
            return librosa.note_to_midi(note)
        except:
            return np.nan

# Apply the conversion to the manual DataFrame
manual_df['MIDI_NOTE_MANUAL'] = manual_df['LABEL_MANUAL'].apply(note_to_midi)

# Display the updated manual DataFrame
display(manual_df)
```

```python
# Merge the manual DataFrame with the automatic pitch analysis DataFrame
merged_df = pd.merge(df, manual_df, on='TIME', how='inner')

# Display the merged DataFrame
display(merged_df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the plot
plt.figure(figsize=(14, 8))

# Plot automatic pitch detections
methods = ['YIN', 'PIPTRACK', 'CREPE', 'PARS']
markers = {'YIN': 'o', 'PIPTRACK': 'x', 'CREPE': 's', 'PARS': 'd'}
colors = {'YIN': 'blue', 'PIPTRACK': 'green', 'CREPE': 'red', 'PARS': 'purple'}

for method in methods:
    midi_col = f'MIDI_NOTE_{method}'
    plt.scatter(
        merged_df['TIME'],
        merged_df[midi_col],
        marker=markers[method],
        color=colors[method],
        label=f'{method} Auto',
        s=100,
        alpha=0.7
    )

# Plot manual pitches
plt.scatter(
    manual_df['TIME'],
    manual_df['MIDI_NOTE_MANUAL'],
    marker='*',
    color='black',
    label='Manual',
    s=200,
    edgecolors='white'
)

# Formatting the y-axis to show note names
# Determine the range of MIDI notes
min_midi = int(min(
    merged_df[['MIDI_NOTE_YIN', 'MIDI_NOTE_PIPTRACK', 'MIDI_NOTE_CREPE', 'MIDI_NOTE_PARS']].min().min(),
    manual_df['MIDI_NOTE_MANUAL'].min()
)) - 1
max_midi = int(max(
    merged_df[['MIDI_NOTE_YIN', 'MIDI_NOTE_PIPTRACK', 'MIDI_NOTE_CREPE', 'MIDI_NOTE_PARS']].max().max(),
    manual_df['MIDI_NOTE_MANUAL'].max()
)) + 1

yticks = np.arange(min_midi, max_midi + 1)
ylabels = [librosa.midi_to_note(midi) for midi in yticks]
plt.yticks(yticks, ylabels)

# Labels and Title
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Note', fontsize=14)
plt.title('Piano Roll Comparison: Manual vs. Automatic Pitch Detection', fontsize=16)

# Legend
plt.legend(fontsize=12)

# Grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
